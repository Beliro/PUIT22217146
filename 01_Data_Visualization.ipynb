```python
# Cell 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# This command makes sure the charts appear directly inside the notebook
%matplotlib inline

print("Libraries imported successfully!")
```

```python
# Cell 2: Load the data
file_path = 'MinoAI_dataset.xlsx - MinoAI dataset.csv'
df = pd.read_csv(file_path)

# Display the first 10 rows
df.head(10)
```

```python
# Cell 3: General Info
# This tells us how many rows we have and if any columns have "null" (missing) values
df.info()

# This gives us a statistical summary (Average price, min/max price, etc.)
df.describe()
```

```python
# Cleaning missing values for visualization
# If reviews_per_month is missing, we assume it's 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Drop rows where critical info like 'name' or 'host_name' is missing
df.dropna(subset=['name', 'host_name'], inplace=True)

print("Missing values handled.")
```

```python
# To check the Price Distribution
plt.figure(figsize=(10, 6))

# We filter out extreme luxury rentals (> $500) for a clearer chart
affordable_df = df[df['price'] < 500] 

sns.histplot(affordable_df['price'], bins=50, kde=True, color='green')
plt.title('Distribution of Prices (Under $500)')
plt.xlabel('Price per Night ($)')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# Here, we are trying to count listings by Neighbourhood Group
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='neighbourhood_group', palette='viridis')
plt.title('Which Borough has the most listings?')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.show()
```

```python
# Cell 7: Map of Listings
plt.figure(figsize=(10, 8))

# We use a scatter plot. X is longitude, Y is latitude.
# We color the dots based on 'neighbourhood_group'
sns.scatterplot(data=affordable_df, x='longitude', y='latitude', hue='neighbourhood_group', alpha=0.5)

plt.title('Map of NYC Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Borough')
plt.show()
```
